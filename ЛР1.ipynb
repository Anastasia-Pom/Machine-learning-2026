```python
Для исследования я выбрала датасет "wine". Предсказания моделей, обучившихся на данном датасете,
могут быть использованы изготовителями вин для прогнозирования определенных показателей продукта,
быть применены людьми, работающими с разными видами вин и их характеристиками. Датасет содержит кололнки:
alcohol
malic_acid
ash
alcalinity_of_ash
magnesium
total_phenols
flavanoids
nonflavanoid_phenols
proanthocyanins
color_intensity
hue
od280/od315_of_diluted_wines
proline
target

Датасет содержит 178 строк и 14 колонок. Все значения в ячейках являются числовыми (float64).
У целевого признака "target" есть три значения (0,1,2)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import *
%matplotlib inline
sns.set(style="ticks")
def make_dataframe(ds_function):
    ds = ds_function()
    df = pd.DataFrame(data= np.c_[ds['data'], ds['target']],
                     columns= list(ds['feature_names']) + ['target'])
    return df
data = make_dataframe(load_wine)
data.head()
data.shape
data.columns
data.dtypes
for col in data.columns:
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
data.describe()
data['target'].unique()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(ax=ax, x='magnesium', y='color_intensity', data=data)
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(ax=ax, x='alcohol', y='proline', data=data)
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(ax=ax, x='alcohol', y='proline', data=data, hue='target')

fig, ax = plt.subplots(figsize=(10,10))
sns.distplot(data['alcohol'])
sns.distplot(data['color_intensity'])
```
